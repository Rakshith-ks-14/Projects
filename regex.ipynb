## 2
    2) write a program to extract all mobile numbers present in file input.txt where
    numbers are mixed with normal text data and write all extracted mobile numbers into output.txt


```python
import requests
from bs4 import BeautifulSoup

u=requests.get("https://www.redbus.in/info/contactus")
c = u.content
soup = BeautifulSoup(c)
soup
numbers=re.findall("[7-9]\d{9}",str(soup))
for n in numbers:
    print(n)
```

```python
import re,urllib
import urllib.request

u=urllib.request.urlopen("http://www.testyantra.com/")
text=u.read()
numbers=re.findall("[0-9-]{7}[0-9-]+",str(text),re.I)
for n in numbers:
    print(n)
```

```python
def add(ar1,ar2):
    s=ar1
    lis=[]
    if ar1>ar2:
        s=ar2
    l=len(ar1)-len(ar2)
    for i in range(l):
        s.insert(0,0)
    for j,k in zip(ar1,ar2):
        lis.append(j+k)
    return lis
add([2,3,4], [1,2])
```
